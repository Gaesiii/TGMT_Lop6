```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
my_path='C:/Users/L7pho/anhtest/test1.jpg'
image = Image.open(my_path)
image_array = np.array(image.convert("L"))
inverted_image = 255 - image_array
squared_image = image_array ** 2
clamped_image = np.clip(image_array, 100, 200)
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)
axs[0, 0].imshow(image_array, cmap="gray")
axs[0, 0].set_title('Hinh anh goc')
axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Hinh anh da bien doi')
axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Hinh anh da bien doi')
axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Hinh anh da bien doii')
plt.show()
```
