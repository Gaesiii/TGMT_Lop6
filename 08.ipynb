```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'bo-', label='du lieu')
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu-1')
plt.legend() 
plt.xlabel('Truc x') 
plt.ylabel('Truc y')
plt.show() 
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color='g', alpha=0.6) 
plt.xlabel('Danh muc') 
plt.ylabel('Gia tri') 
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Bieu do hinh tron')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Du Lieu')  
plt.title('Bieu do phan tan vi du')  
plt.xlabel('Truc x') 
plt.ylabel('Truc y')  
plt.legend()  
plt.show()  
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  
plt.title('Bieu do hop vi du')  
plt.ylabel('Gia tri') 
plt.show() 
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data) 
plt.title('Bieu doi violin vi du')  
plt.ylabel('Gia tri') 
plt.show()  
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Bieu do dam may tu van ban vi du')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Bieu do radar vi du')
plt.show()
```
